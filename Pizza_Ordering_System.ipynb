```python
TAX_RATE = .08  # Sales tax rate
discount = False  # Flag to track if user is eligible for a discount
number_of_toppings = 0  # Counter for the number of additional toppings
toppings = "Cheese"  # Default topping is "Cheese"

def main():
    # Prompt user and get first name.
    print("Welcome to Mike and Diane's Pizza")
    first_name = input("Enter your first name: ")

    # ADD LINES HERE FOR TASK #1
    # Determine if user is eligible for discount by
    # having the same first name as one of the owners.
    if first_name.lower() == "mike" or first_name.lower() == "diane":
        discount = True  # Set discount flag to True for eligible users

    # Prompt user and get pizza size choice.
    print("\nPizza Size (inches)   Cost\n" +
          "        10            $10.99\n" +
          "        12            $12.99\n" +
          "        14            $14.99\n" +
          "        16            $16.99\n")
    inches = int(input("What size pizza would you like?" +
                       "\n 10, 12, 14, or 16 (enter the number only): "))

    # Set price and size of pizza ordered.
    cost = 12.99  # Default cost for 12-inch pizza
    if inches == 10:
        cost = 10.99
    elif inches == 12:
        cost = 12.99
    elif inches == 14:
        cost = 14.99
    elif inches == 16:
        cost = 16.99
    else:
        inches = 12  # If an invalid size is entered, set size to 12 inches
        cost = 12.99
        print("That was not one of the choices, you will get a 12 inch pizza")

    # Prompt user and get crust choice.
    print("\nWhat type of crust " +
          "do you want? ")
    crust_type = input("(H) Hand-tossed, \n" +
                       "(T) Thin-crust, or \n" +
                       "(D) Deep-dish \n" +
                       "(enter H, T, or D): ")

    # Set user's crust choice on pizza ordered.
    if crust_type == 'h' or crust_type == "H":
        crust = "Hand-tossed"
    elif crust_type == 't' or crust_type == "T":
        crust = "Thin-crust"
    elif crust_type == 'd' or crust_type == "D":
        crust = "Deep-dish"
    else:
        print("That was not one of " +
              "the choices, you will " +
              "get Hand-tossed.")
        crust = "Hand-tossed"

    # Prompt user and get topping choices one at a time.
    toppings = "Cheese"  # Default topping is "Cheese"
    number_of_toppings = 0  # Reset the topping counter
    print("\nAll pizzas come with cheese.")
    print("Additional toppings are $1.25 each, choose from:\n \tPepperoni, Sausage, Onion, Mushroom")
    choice = input("Do you want Pepperoni? (Y/N): ")

    if choice == 'Y' or choice == "y":
        toppings = toppings + " Pepperoni"
        number_of_toppings += 1

    choice = input("Do you want Sausage? (Y/N): ")
    if choice == 'Y' or choice == "y":
        toppings = toppings + " Sausage"
        number_of_toppings += 1

    choice = input("Do you want Onion? (Y/N): ")
    if choice == 'Y' or choice == "y":
        toppings = toppings + " Onion"
        number_of_toppings += 1

    choice = input("Do you want Mushroom? (Y/N): ")
    if choice == 'Y' or choice == "y":
        toppings = toppings + " Mushroom"
        number_of_toppings += 1

    # Add additional toppings cost to the cost of the pizza.
    cost = cost + (1.25 * number_of_toppings)

    # Display order confirmation.
    print("\nYour order is as follows: ")
    print(inches, "inch pizza")
    print(crust, "crust")
    print("Toppings:", toppings)

    # Apply discount if user is eligible.
    if discount:
        print("You are eligible for a $2 discount")
        cost -= 2

    # Print the cost of the order
    print(f"The cost of your order is: {cost:.2f}")

    # Calculate and display tax and total cost.
    tax = cost * TAX_RATE
    print(f"The tax is: {tax:.2f}")
    total_cost = tax + cost
    print(f"The total due is {total_cost:.2f}")

    print("\nYour order will be ready for pickup in 30 minutes.")


if __name__ == "__main__":
    main()
```
